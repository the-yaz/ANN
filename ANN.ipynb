```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
x=dataset.iloc[:, 3:13].values
y=dataset.iloc[:, 13].values

#visualisation du resultat pour comprendre
q=pd.DataFrame(x)
print(q)
q1=pd.DataFrame(y)
print(q1)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#coudage des valeurs catigoriel
labeloncoder_X_1 = LabelEncoder()
x[:, 1]=labeloncoder_X_1.fit_transform(x[:, 1])
labeloncoder_X_2 = LabelEncoder()
x[:, 2]=labeloncoder_X_2.fit_transform(x[:, 2])

#visualisation du resultat pour comprendre
q=pd.DataFrame(x)
print(q)
```

```python
#pour eviter la clasification entre les pays
ct = ColumnTransformer([("Geography", OneHotEncoder(), [1])],remainder='passthrough')
x=ct.fit_transform(x)
x=x[: , 1:]
#visualisation du resultat pour comprendre
q=pd.DataFrame(x)
print(q)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train, y_test =train_test_split(x,y,test_size =0.2, random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x=sc.fit_transform(x)
#visualisation du resultat pour comprendre
q=pd.DataFrame(x_train)
print(q)
q1=pd.DataFrame(x_test)
print(q1)
```

#Création de notre réseau ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Etape 1 : Initialistion du ANN
classifier = Sequential()
#Ajouter la couche d'entrée et une couche cachée
classifier.add(Dense(units=6 , activation="relu" , kernel_initializer="uniform",input_dim=11))
#classifier.add(Dropout(rat=0.1))
#ajouter une deuxiem couche cachée
classifier.add(Dense(units=6 , activation="relu" , kernel_initializer="uniform"))
#classifier.add(Dropout(rat=0.1))
#ajouter une deuxiem couche de sortie
classifier.add(Dense(units=1 , activation="sigmoid" , kernel_initializer="uniform"))
```

```python
#Compilation
classifier.compile(optimizer="adam", loss="binary_crossentropy",metrics=["accuracy"])                                           
#Entrainement du reseau
classifier.fit(x_train,y_train, batch_size=10, epochs=100)
```

```python
y_pred= classifier.predict(x_test)
y_pred= (y_pred > 0.5)
#confusion matrix

from sklearn.metrics import confusion_matrix
cn = confusion_matrix(y_test , y_pred)
print(cn)
print(y_pred)
```

```python

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def build_classifier():

  classifier=Sequential()
  #Ajouter la couche d'entrée et une couche cachée
  classifier.add(Dense(units=6 , activation="relu" , kernel_initializer="uniform",input_dim=11))
  #classifier.add(Dropout(rat=0.1))
  #ajouter une deuxiem couche cachée
  classifier.add(Dense(units=6 , activation="relu" , kernel_initializer="uniform"))
  #classifier.add(Dropout(rat=0.1))
  #ajouter une deuxiem couche de sortie
  classifier.add(Dense(units=1 , activation="sigmoid" , kernel_initializer="uniform"))
  #Compilation
  classifier.compile(optimizer="adam", loss="binary_crossentropy",metrics=["accuracy"])
  return classifier
```

```python

classifier= KerasClassifier(build_fn=build_classifier, batch_size=10, epochs=100)
Precision = cross_val_score(estimator = classifier,X=x,y=y, cv=11)
moyenne = Precision.mean()
ecart_type= Precision.std()

print(Precision)
print(moyenne)
print(ecart_type)
```
